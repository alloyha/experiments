```python
import pytest
import ipytest
import json

ipytest.autoconfig()
```

```python
delimiter = ','
def unwrap(str_):
    return str_.split(delimiter)

def inverse_dict(dict_):
    _dict = {}
    
    for key, values in dict_.items():
        for value in values:
            _dict[str(value)] = unwrap(key)
    
    return _dict

def flatten(l):
    return [item for sublist in l for item in sublist]

def json_prettify(json_):
    return json.dumps(json_, indent=4);
```

```python
%%ipytest

# define the tests
def test_unwrap():
    assert unwrap('1,2,3') == ['1','2','3']

def test_inverse_dict():
    assert inverse_dict(
        {'a': [1, 2], 'b': [3, 4]}
    ) == {
            '1': ['a'], 
            '2': ['a'],
            '3': ['b'],
            '4': ['b']
         }
    
def test_inverse_dict():
    assert inverse_dict(
        {'a': [1, 2], 'b': [3, 4]}
    ) == {
            '1': ['a'], 
            '2': ['a'],
            '3': ['b'],
            '4': ['b']
         }

def test_flatten():
    assert flatten([[1, 2], [3, 4]]) == [1, 2, 3, 4]
    
def test_json_prettify():
    assert json_prettify({'a': 1, 'b': 2}) == '{\n    "a": 1,\n    "b": 2\n}'
```

```python
import json

with open('./apache_products.json') as f:
    products = json.load(f)

print(json_prettify(products))
```

```python
from eule import euler

# Pizza to exclusive ingredient
euler_products = euler(products);

print(json_prettify(euler_products));
```

```python
from json import loads 

category_to_product=loads(json_prettify(inverse_dict(euler_products)))

category_euler=euler(category_to_product)

for category_item in category_euler.items():
    print(category_item[0]+': '+str(category_item[1]))

products_check=[]
for category_products in category_euler.values():    
    products_check = products_check+category_products
    
print(len(products_check))
```
