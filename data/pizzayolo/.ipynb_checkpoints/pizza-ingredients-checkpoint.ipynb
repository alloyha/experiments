```python
import pytest
import ipytest
import json

ipytest.autoconfig()
```

```python
delimiter = ','
def unwrap(str_):
    return str_.split(delimiter)

def inverse_dict(dict_):
    _dict = {}
    
    for key, values in dict_.items():
        for value in values:
            _dict[str(value)] = unwrap(key)
    
    return _dict

def flatten(l):
    return [item for sublist in l for item in sublist]

def json_prettify(json_):
    return json.dumps(json_, indent=4);
```

```python
%%ipytest

# define the tests
def test_unwrap():
    assert unwrap('1,2,3') == ['1','2','3']

def test_inverse_dict():
    assert inverse_dict(
        {'a': [1, 2], 'b': [3, 4]}
    ) == {
            '1': ['a'], 
            '2': ['a'],
            '3': ['b'],
            '4': ['b']
         }

def test_flatten():
    assert flatten([[1, 2], [3, 4]]) == [1, 2, 3, 4]
    
def test_json_prettify():
    assert json_prettify({'a': 1, 'b': 2}) == '{\n    "a": 1,\n    "b": 2\n}'
```

```python
import json

with open('./pizzas.json') as f:
    pizzas = json.load(f)

print(json_prettify(pizzas))
```

```python
from eule import spread_euler

ingredients = list(set(flatten(list(pizzas.values()))))
print(ingredients);

pizza_to_ingredients = spread_euler(pizzas);
print(json_prettify(pizza_to_ingredients));

ingredients_to_pizza = inverse_dict(pizza_to_ingredients);
print(json_prettify(ingredients_to_pizza));

print(
    dict(
        sorted(
            {
                ingredients: len(pizzas) 
                for ingredients, pizzas in ingredients_to_pizza.items()
            }.items(), key=lambda item: item[1]
        )
    )    
)
```

```python
import json
import os

source_folder = './baked_pizzas/'
pizzarias = {}

for fname in os.listdir(source_folder):
    pizza_key = fname.split('.')[0].split('_')[0]
    
    with open(source_folder+fname) as f:
        pizzarias[pizza_key] = json.load(f)

pizzarias_flavors = {name: list(pizzas.keys()) for name, pizzas in pizzarias.items()}

pizzarias_ingredients = {
    name: list(set(flatten(list(pizzas.values())))) 
    for name, pizzas in pizzarias.items()
}

```

```python
pizzarias_to_flavors = spread_euler(pizzarias_flavors);
print(json_prettify(pizzarias_to_flavors));
```

```python
pizzarias_to_ingredients = spread_euler(pizzarias_ingredients);
print(json_prettify(pizzarias_to_ingredients));
```

```python
ingredient_to_pizzarias = inverse_dict(pizzarias_to_ingredients);
print(json_prettify(ingredient_to_pizzarias));
```

```python
print(
    dict(
        sorted(
            {
                ingredients: len(pizzas) 
                for ingredients, pizzas in ingredient_to_pizzarias.items()
            }.items(), key=lambda item: item[1]
        )
    )    
)
```
