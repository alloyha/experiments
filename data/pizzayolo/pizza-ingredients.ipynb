```python
import pytest
import ipytest
import json

ipytest.autoconfig()
```

```python
delimiter = ','
def unwrap(str_):
    return str_.split(delimiter)

def inverse_dict(dict_):
    _dict = {}
    
    for key, values in dict_.items():
        for value in values:
            _dict[str(value)] = unwrap(key)
    
    return _dict

def flatten(l):
    return [item for sublist in l for item in sublist]

def json_prettify(json_):
    return json.dumps(json_, indent=4)
```

```python
%%ipytest

# define the tests
def test_unwrap():
    assert unwrap('1,2,3') == ['1','2','3']

def test_inverse_dict():
    assert inverse_dict(
        {'a': [1, 2], 'b': [3, 4]}
    ) == {
            '1': ['a'], 
            '2': ['a'],
            '3': ['b'],
            '4': ['b']
         }

def test_flatten():
    assert flatten([[1, 2], [3, 4]]) == [1, 2, 3, 4]
    
def test_json_prettify():
    assert json_prettify({'a': 1, 'b': 2}) == '{\n    "a": 1,\n    "b": 2\n}'
```

```python
import json

with open('./pizzas.json') as f:
    pizzas = json.load(f)

print(json_prettify(pizzas))
```

```python
# Available ingredients 
ingredients = list(set(flatten(list(pizzas.values()))))
print(ingredients);
```

```python
from eule import euler

# Pizza to exclusive ingredient
pizza_to_ingredients = euler(pizzas);
print(json_prettify(pizza_to_ingredients));
```

```python
# Ingredient to pizza
ingredients_to_pizza = inverse_dict(pizza_to_ingredients);
print(json_prettify(ingredients_to_pizza));
```
